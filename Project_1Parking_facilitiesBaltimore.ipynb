```python
cleanedMaxRate=df['dailyMaximumRate'].str.extract(r'^\D*(\d+)', expand=False)
cleanedMaxRate

#cleanedMaxRate=df['dailyMaximumRate'].str.extract(r'^\D*(\d+)', expand=False)
```

```python
 # References -  https://realpython.com/python-data-cleaning-numpy-pandas/

df['dailyMaximumRate'] = pd.to_numeric(cleanedMaxRate)
df['dailyMaximumRate'].dtype
```

```python
df['dailyMaximumRate'].describe()
```

```python
plt.hist(df['dailyMaximumRate'], bins=12,)
plt.title('Maximum Daily Rates')
plt.xlabel('maximum daily parking charges($)')
plt.ylabel('Frequency')
plt.show()
```

```python
df['dailyMaximumRate']
```

```python
cleanMonthRate=df['monthlyRate'].str.extract(r'^\D*(\d+)', expand=False)
cleanMonthRate
```

```python
df['monthlyRate'] = pd.to_numeric(cleanMonthRate)
df['monthlyRate'].dtype
```

```python
df['monthlyRate'].describe()
```

```python
plt.hist(df['monthlyRate'], bins=10)
plt.title('Monthly charges')
plt.xlabel('Monthly parking rates($)')
plt.ylabel('Frequency')
plt.show()
```

```python
df.plot.scatter(y='dailyMaximumRate',x='monthlyRate',c='Blue',figsize=(10,8))
plt.title('Monthly rates against Maximum daily rates')
plt.xlabel('Monthly parking rates($)')
plt.ylabel('Maximum daily rates($)')
```

We can infer from the scatter plot that, Although the maximum daily parking charges are higher for in certain areas their corresponding monthly charges are in the range of 150 to 200 dollars. 
So it is better to buy a monthly pass as nearly half of paying it daily.
